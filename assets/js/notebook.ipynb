```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import requests
import json
```

```python
btc1m = requests.get('https://ftx.com/api/markets/BTC-PERP/candles?resolution=3600').json()
btc1m = pd.DataFrame(btc1m['result'])
btc1m.drop(['startTime'], axis = 1, inplace=True)
btc1m['time'] = pd.to_datetime(btc1m['time'], unit='ms')
btc1m.set_index('time', inplace=True)
btc_df = btc1m.drop(['high', 'low', 'close', 'volume'],axis=1)
btc_df.rename(columns ={'open':'btcopen'},inplace=True)
btc_df['btc'] = 'BTC'
# btc_df.reset_index(level=0, inplace=True)

btc_df.head(20)
```

```python
sol = requests.get('https://ftx.com/api/markets/SOL/USD/candles?resolution=3600').json()
sol = pd.DataFrame(sol['result'])
sol.drop(['startTime'], axis = 1, inplace=True)
sol['time'] = pd.to_datetime(sol['time'], unit='ms')
sol.set_index('time', inplace=True)
sol_df = sol.drop(['high', 'low', 'close', 'volume'],axis=1)
sol_df['sol'] = 'SOL'
sol_df.rename(columns ={'open':'solopen'},inplace=True)



sol_df.head(20)
```
